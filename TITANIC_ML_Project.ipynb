```python
#Mapping Text_Data to Binary Value
income_set = set(df['sex'])
df['sex'] = df['sex'].map({'male': 0,'female': 1})
df.head
```

```python
#Segregation of Dataset into X and Y

X = df.drop('survived', axis = 'columns')
X
```

```python
Y = df.survived
Y
```

```python
# Segregation of Dataset into X and Y Variables
X.columns[X.isna().any()]
X.age = X.age.fillna(X.age.mean())
```

```python
#Finding NA values from the features X
X.columns[X.isna().any()]
```

```python
#Removing NA values from the features X
X.fare = X.fare.fillna(X.fare.mean())
```

```python
#Finding NA values from the features X again
X.columns[X.isna().any()]
```

```python
# Splitting  Dataset in Train and Test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split (X, Y, test_size = 0.25, random_state=0)
```

```python
# Training using Algorithm
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit (X_train, y_train)
```

```python
# Predicting whether Person Survived or Not
pclassNo = int(input("Enter Person's Pclass number: "))
gender = int(input("Enter Persons's Gender 0-female 1-male(0 or 1):"))
age = int(input("Enter Person's Age : "))
fare = float(input("Enter Person's Fare:"))
person = [[pclassNo, gender, age, fare]]
result = model.predict(person)
print(result)

if result ==1:
    print("Person might be survived")
else:
    print("Person might not be survived")
```

```python
# Prediction for all Test Data

y_pred = model.predict(X_test)
print(np.column_stack((y_pred,y_test)))
```

```python
# Accuracy of the Model
from sklearn.metrics import accuracy_score
print ("Accuracy of the Model:{0}%".format(accuracy_score(y_test, y_pred)*100))
```
